```python
import os
import requests
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import json
from typing import List
from keplergl import KeplerGl
```

```python
# issues here
GOLEMIO_KEY = os.environ.get('GOLEMIO_KEY')
```

```python
def get_golemio(data:str, golemio_key):
    url = 'https://api.golemio.cz/' + data
    headers = {"X-Access-Token": golemio_key,
            'Accept': 'application/json'}

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Failed to retrieve data: {response.status_code}")
        print(response.text)
```

```python
bicyclecounters = get_golemio('v2/bicyclecounters?latlng=50.124935%2C14.457204&range=5000&limit=100&offset=0', GOLEMIO_KEY)
bicyclecounters_df = pd.DataFrame.from_dict(bicyclecounters)
bicyclecounters_df.head()
```

```python
bicyclecounters_df = gpd.GeoDataFrame.from_features(bicyclecounters_df.iloc[:,0], crs="EPSG:4326")
```
