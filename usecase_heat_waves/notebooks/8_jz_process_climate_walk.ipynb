```python
import pandas as pd
import plotly.express as px
```

```python
survey_df = pd.read_csv("../data/raw/climate walk/survey_0.csv", index_col="CreationDate")
survey_df.index = pd.to_datetime(survey_df.index)
survey_df.drop(["Shape__Area", "Shape__Length", "EditDate", "Creator", "Editor", "Rate your Thermal Comfort right now", "Describe this spot"], 
               axis=1,
               inplace=True)

location = ['In_direct_sun',
            "In_the_shade",
            'Cloudy', 
            'Windy',
            'No_Wind',
            "Good_Air_Quality",
            'Poor_Air_Quality',
            "Some_Background_Noise",
            'Quiet', 
            'Distracting_Noise'
]
survey_df["How is this spot relative to previous spots"] = survey_df["How is this spot relative to previous spots"].str.replace("Strongly disagree", "Much Cooler")
for l in location:
    survey_df[l] = survey_df["Describe this spot.1"].str.contains(l)

survey_df = survey_df.loc[survey_df.index>"17.09.2024"]
survey_df.head()
```
